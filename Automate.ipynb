```python
dir='E:\\Extra\\Workplace\\Projects\\rameshbisht.github.io-master\\res'
```

```python
data=[]
for x in range(36):
    dir1=os.path.join(dir,str(x+1))
    obj={}
    obj['thumbnail']=os.path.join('res',str(x+1),'1.jpg').replace('\\\\','\\')
    obj['title']='loral epsum'
    slides=[]
    for y in os.listdir(dir1):
        file1=os.path.join(dir1,y)
        slide={}
        img=cv2.imread(file1)
        h,w,c=img.shape
        slide['src']=os.path.join('res',str(x+1),str(y))
        slide['w']=w
        slide['h']=h
        slides.append(slide)
        print(os.path.join('res',str(x+1),str(y)))
    obj['slides']=slides
    data.append(obj)
```

```python
with open('data.json', 'w') as outfile:
    json.dump(data, outfile)
```
